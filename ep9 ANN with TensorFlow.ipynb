```python
import numpy as np
from random import random
from sklearn.model_selection import train_test_split

def generate_dataset(num_samples,test_size):
    # create dummy data
    x = np.array([[random()/2 for _ in range(2)] for _ in range(num_samples)])
    y = np.array([[i[0]+i[1]] for i in x])
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=test_size)
    return x_train, x_test, y_train, y_test
```

```python
x_train, x_test, y_train, y_test = generate_dataset(5000,.2)
print("x_test:  {}  x_train:  {}".format(x_test.shape,x_train.shape))
print("y_test:  {}  y_train:  {}".format(y_test.shape,y_train.shape))
```

```python
import tensorflow as tf

# build model   2 -> 5 -> 1
model = tf.keras.Sequential([
    tf.keras.layers.Dense(5, input_dim=2, activation='sigmoid'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# compile model
optimizer = tf.keras.optimizers.SGD(learning_rate=0.1)
model.compile(optimizer=optimizer,loss='MSE')
# train model
model.fit(x_train,y_train,epochs=100,batch_size=20)
# evaluate model
print("\n model evaluatoin:")
model.evaluate(x_test,y_test,verbose=1)
```

## prediction


```python
data = np.array([[0.1,0.2],[0.2,0.2]])
expected_output = np.array([[0.3],[0.4]])
predictions = model.predict(data)

print("\n Predictions")
for d , p in zip(data,predictions):
    print("{} + {} = {}".format(d[0],d[1],p[0]))
```
